```python
% matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
#3.2 Creating data array and label vector
data=[]
label=[]
with open('iris.data','rt') as csv_file:
    allsamples = csv.reader(csv_file, delimiter="\t")
    for sample in allsamples:
        if sample:
            sample = sample[0]
            sample = sample.split(",")
            data.append(sample[:-1])
            label.append(sample[len(sample)-1])
data = np.array(data)
data=data.astype(float)
# print(data)
# print(label)
```

```python
#3.3 Visualize dataset
color_label = [l.replace('Iris-setosa', 'r') for l in label]
color_label = [l.replace('Iris-versicolor', 'g') for l in color_label]
color_label = [l.replace('Iris-virginica', 'b') for l in color_label]
```

```python
plt.scatter (data[:,0],data[:,1],c = color_label)
plt.title("Iris Flowers")
plt.xlabel("sepal length in cm")
plt.ylabel("sepal width in cm")
#plt.savefig ("plot1.png")
```
